```python
###############
# Fourier series of discrete time periodic signal
###############

import numpy as np
import matplotlib.pyplot as plt

time = np.arange(0, 1, 0.0001)
frequency = 100
N = 1024
signal = np.sin(2 * np.pi * frequency * time)

print(signal[1:N+1])
plt.plot(signal[1:N+1])
```

```python
###############
# Fourier analysis
###############

x = signal
C = np.zeros(N)

# calculating fourier coefficient C[k]
for k in range(N):
  for n in range(N):
    C[k] += (x[n] * np.exp(-2j * np.pi * k * n/N))
  C[k] /= N

plt.plot(np.abs(C)**2)
```

```python
##############
# Fourier synthesis
##############

x = np.zeros(N)

# calculating fourier synthesis values x[n]
for n in range(N):
  for k in range(N):
    x[n] += (C[k] * np.exp(2j * np.pi * k * n/N))

plt.plot(x[1:N+1])

```
