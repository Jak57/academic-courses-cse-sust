```python
##############
# Compute DFT of a signal using DFT equation
##############

import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0, 1, 0.0001)
f = 100
N = 1024
x = np.sin(2 * np.pi * f * t) + np.sin(2 * np.pi * 385 * t) + np.sin(2 * np.pi * 690 * t)

plt.plot(x[1:N+1])
```

```python
X = np.zeros(N)

# computing DFT X[k]
for k in range(N):
  for n in range(N):
    X[k] += (x[n] * np.exp(-2j * np.pi * k * n/N))

plt.plot(np.abs(X)**2)
```
