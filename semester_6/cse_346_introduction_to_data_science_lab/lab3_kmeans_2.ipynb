# Scaling

```python
new_df = df[['SepalLengthCm', 'SepalWidthCm']]
new_df
```

```python
scaler = MinMaxScaler()
norm_array = scaler.fit_transform(new_df)
```

```python
df = pd.DataFrame(norm_array, columns=new_df.columns)
df
```

# KMeans

```python
km = KMeans(n_clusters=3)
y_pred = km.fit_predict(df)
y_pred
```

```python
df['cluster'] = y_pred
```

```python
df.head()
```

```python
km.cluster_centers_
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1['SepalLengthCm'], df1['SepalWidthCm'],color='green')
plt.scatter(df2['SepalLengthCm'], df2['SepalWidthCm'],color='red')
plt.scatter(df3['SepalLengthCm'], df3['SepalWidthCm'],color='black')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.legend()
```
